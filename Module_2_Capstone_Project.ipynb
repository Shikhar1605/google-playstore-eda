In the Play Store dataset, some columns contain 0 values. It is important to check whether these 0s are meaningful or just incorrect data.

**Reviews** = 0

This is valid, because some apps may not have received any reviews yet. Therefore, we keep these rows.

**Price** = 0

This means the app is free, which is correct for most apps. We keep these values.

**Installs** = 0

This is unrealistic because any app on the Play Store must have at least 1 install. Therefore, rows with Installs = 0 are considered invalid and are dropped.

**Size** = 0

This is also unrealistic because an app cannot have a size of 0 MB. Therefore, we replace 0 with NaN so that it is treated as missing data.

By doing this, we ensure that our dataset only contains valid and meaningful values for analysis.

````python
# Dropping rows where Installs = 0 (not valid for Play Store apps)
before_rows = play_store.shape[0]
play_store = play_store[play_store['Installs'] != "0"]   # 'Installs' is string type originally
after_rows = play_store.shape[0]
print(f"Dropped {before_rows - after_rows} rows where Installs = 0")

# Convert Size column to numeric (MB) and replace 0 with NaN
# Cleaning size column
def convert_size(size):
    if isinstance(size, str):
        size = size.strip()
        if size.endswith("M"):
            return float(size[:-1])   # in MB
        elif size.endswith("k") or size.endswith("K"):
            return float(size[:-1]) / 1024  # KB to MB
        elif size == "Varies with device":
            return None
    return None

# Applying conversion
if play_store['Size'].dtype == object:
    play_store['Size_MB'] = play_store['Size'].apply(convert_size)

# Replace 0 with NaN(Not a Number) in Size_MB
play_store['Size_MB'] = play_store['Size_MB'].replace(0, np.nan)

print("Converted Size column to MB and replaced 0 with NaN")
````

````python
# Replace 0 in Type column with "Free"
play_store['Type'] = play_store['Type'].replace("0", "Free")
print("Replaced 0 in Type column with 'Free'")
````

````python
# Check how many 0 values each numeric column has
zero_counts = (play_store == "0").sum() + (play_store == 0).sum()
print(f"Count of 0 values in Play Store dataset:\n{zero_counts}")
````

### Changing Data Type

````python
# Check column info (including data types) for Play Store dataset
print("Play Store dataset info:")
play_store.info()

# Check column info (including data types) for User Reviews dataset
print("\nℹUser Reviews dataset info:")
user_reviews.info()
````

````python
# Reviews are stored as strings. We convert to integers for numeric analysis.
# Convert Reviews to integer
play_store['Reviews'] = pd.to_numeric(play_store['Reviews'], errors='coerce')
print(f"Reviews converted to int — dtype is now {play_store['Reviews'].dtype}")
````

````python
# Installs are strings like "1,000+"; converting to integers allows numeric comparisons.
# Clean and convert Installs to integer
play_store['Installs_clean'] = (
    play_store['Installs']
    .str.replace('+','', regex=False)
    .str.replace(',','', regex=False)
)
play_store['Installs_clean'] = pd.to_numeric(play_store['Installs_clean'], errors='coerce')
print(f"Installs converted to int — dtype is now {play_store['Installs_clean'].dtype}")
````

````python
# Price is stored as "$2.99" or "0"; converting allows analysis of paid vs free apps.
# Clean and convert Price to float
play_store['Price_clean'] = (
    play_store['Price']
    .str.replace('$','', regex=False)
    .str.replace(',','', regex=False)
)
play_store['Price_clean'] = pd.to_numeric(play_store['Price_clean'], errors='coerce')
print(f"Price converted to float — dtype is now {play_store['Price_clean'].dtype}")
````

````python
# Size is in strings like "19M" or "870k". We convert into numeric MB values for analysis.
# Convert Size to MB
def convert_size(size):
    if isinstance(size, str):
        size = size.strip()
        if size.endswith("M"):
            return float(size[:-1])
        elif size.endswith("k") or size.endswith("K"):
            return float(size[:-1]) / 1024
        elif size == "Varies with device":
            return None
    return None
````

````python
# Ratings should be numeric between 1–5. Ensure dtype is float and drop invalid values.
# Convert Rating to float and drop invalid (>5)
play_store['Rating'] = pd.to_numeric(play_store['Rating'], errors='coerce')
invalid_ratings = play_store[play_store['Rating'] > 5].shape[0]
play_store = play_store[play_store['Rating'] <= 5]
print(f"Rating cleaned — dropped {invalid_ratings} invalid rows")
````

````python
# Converting update dates to datetime allows time-based analysis.
# Convert Last Updated to datetime
play_store['Last Updated'] = pd.to_datetime(play_store['Last Updated'], errors='coerce')
print(f"Last Updated converted to datetime — dtype is now {play_store['Last Updated'].dtype}")
````

````python
# Columns like Type, Category, Content Rating, Genres are categorical → reduce memory & enable grouping.
# Convert selected columns to category
categorical_cols = ['Type', 'Category', 'Content Rating', 'Genres']
for col in categorical_cols:
    play_store[col] = play_store[col].astype('category')

print(f"Converted {categorical_cols} to category type")
````

````python
# Confirming all data types are correct
play_store.info()
````

We convert Reviews and Installs_clean from float to integer because these columns represent counts (whole numbers). Using integers makes the dataset more accurate and avoids confusion with decimals.

We also drop the old raw columns (Size, Installs, Price) because they are string-based and no longer needed after creating cleaned numeric versions (Size_MB, Installs_clean, Price_clean). Keeping only the cleaned columns ensures the dataset is tidy and easier to work with during analysis.

````python
# Convert Reviews and Installs_clean to int (since they are counts, not decimals)
play_store['Reviews'] = play_store['Reviews'].astype('int64')
play_store['Installs_clean'] = play_store['Installs_clean'].astype('int64')

print(f"Converted Reviews dtype to {play_store['Reviews'].dtype}")
print(f"Converted Installs_clean dtype to {play_store['Installs_clean'].dtype}")

# Drop old raw columns (string-based)
play_store.drop(['Size', 'Installs', 'Price'], axis=1, inplace=True)

print(f"Dropped raw columns: Size, Installs, Price")
print(f"Final dataset shape: {play_store.shape}")
````

### Addition of New Columns

````python
# Add Is_Free column
# Easier to analyze free vs paid apps.
play_store['Is_Free'] = (play_store['Price_clean'] == 0).astype(int)
print("Added Is_Free column (1 = Free, 0 = Paid)")

# Extract year from Last Updated
# Helps in time-based analysis of updates.
play_store['Last_Updated_Year'] = play_store['Last Updated'].dt.year
print("Added Last_Updated_Year column")

# Create Size_Category
# Makes it easy to compare apps by size groups.
def categorize_size(size):
    if pd.isna(size):
        return "Unknown"
    elif size < 10:
        return "Small"
    elif size < 50:
        return "Medium"
    else:
        return "Large"

play_store['Size_Category'] = play_store['Size_MB'].apply(categorize_size)
print("Added Size_Category column (Small/Medium/Large/Unknown)")

# Estimate Revenue_Potential
# A rough proxy for possible revenue from paid apps.
play_store['Revenue_Potential'] = play_store['Price_clean'] * play_store['Installs_clean']
print("Added Revenue_Potential column")
````

````python
# Chart - 7 visualization code
# Top 10 Categories by Estimated Revenue Potential (Bar Chart)
# Explanation: Revenue_Potential = Price_clean * Installs_clean (estimate)

from matplotlib.ticker import FuncFormatter

# Ensure Revenue_Potential exists; if not, compute it
if 'Revenue_Potential' not in play_store.columns:
    play_store['Revenue_Potential'] = play_store['Price_clean'] * play_store['Installs_clean']

# Group by category and sum revenue potential (observed=True to silence future warnings)
cat_revenue = (
    play_store.groupby('Category', observed=True)['Revenue_Potential']
    .sum()
    .sort_values(ascending=False) )

# Keep only categories with >0 revenue and pick top 10
cat_revenue = cat_revenue[cat_revenue > 0].head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=cat_revenue.values,
    y=cat_revenue.index,
    hue=cat_revenue.index,
    palette="rocket",
    dodge=False,
    legend=False )

# Format x-axis to show human-readable numbers (e.g., 1M, 100K)
def human_format(x, pos):
    if x >= 1_000_000_000:
        return f'{x/1_000_000_000:.1f}B'
    if x >= 1_000_000:
        return f'{x/1_000_000:.1f}M'
    if x >= 1_000:
        return f'{x/1_000:.1f}K'
    return f'{int(x)}'

plt.gca().xaxis.set_major_formatter(FuncFormatter(human_format))

plt.title("Top 10 Categories by Estimated Revenue Potential (Price × Installs)")
plt.xlabel("Estimated Revenue Potential")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart shows where money is most likely to come from by summing Price_clean * Installs_clean per category. That directly links category choice to business value.

##### 2. What is/are the insight(s) found from the chart?

Which categories (e.g., Productivity, Entertainment, Tools) contribute the most estimated revenue. It reveals whether high-install categories are also high-revenue or if some smaller categories punch above their weight because of higher prices.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

* **Positive insight:** If a category shows high revenue potential, devs/businesses can prioritise features/marketing there to accelerate growth.

* **Negative insight:** If revenue is concentrated in a few categories or dominated by a handful of apps, that indicates high dependency/centralization — risky if competition or platform changes occur. Also, because this is an estimate (uses installs × price), it ignores conversion rates and in-app purchases, so interpret cautiously.
